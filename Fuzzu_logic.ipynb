```python
# @title Required Lib
!pip install colorama
from colorama import Fore, Back, Style
import matplotlib.pyplot as plt
import numpy as np
```

```python
# @title Pperceptron network
def HardLimit(ActvationInput) :
  if ActvationInput >= 0:
    return 1
  else :
    return 0

def PerceptronError(Target_Output,Trained_Output) :
  return Target_Output - Trained_Output

def prediction(X,w,b,PRINT=False):
  y=[]
  for i in range(X.shape[0]):
    x = X[i]
    y.append(HardLimit(sum(w*x)+b))
    if(PRINT == True):
      print(f'Summation Opertaion ------>  {sum(w*x)+b} = {w}*{x}+{b}')
      print(f"Applying the activation Function ------> {sum(w*x)+b} = [{HardLimit(sum(w*x)+b)}]")
  return y

def Perceptron (Input, Target,PRINT=False):
  weight = np.random.random(Input.shape[1])
  bias   = np.random.random(1)
  i = 0
  while ( i < Input.shape[0]):
    Y_input = sum(weight*Input[i])
    Trained_Output= HardLimit(Y_input + bias)
    err = PerceptronError(Target[i],Trained_Output)
    bias += err
    if(PRINT == True):
      print("\n----------NEXT-ITER-------------")
      print(f"Weight {weight} Value {i}")
      print(f'Summation Opertaion ------>  {Y_input} = {weight}*{Input[i]}')
      print(f"Applying the activation Function ------> {Y_input} = {HardLimit(Y_input + bias)}")
      print(f"Checking the condion Original : {Target[i]} AND Trained : {Trained_Output}")
      print(Fore.RED +f"Perceptron Error is : {err}"+Style.RESET_ALL)
    if Target[i]!=Trained_Output :
      for j in range(Input.shape[1]):
        weight[j] = weight[j] + err * Input[i][j]
        if(PRINT == True):
          print(f"{weight[j]} = {weight[j]} + {err} * {(Input[i][j])}")
      i=0;
    else :
      i += 1

  return weight , bias
```
